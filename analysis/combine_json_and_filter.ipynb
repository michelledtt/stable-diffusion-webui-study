```python
import json

def combine_json_files(file_paths, output_json_file, output_txt_file):
    # Initialize an empty list to store prompt values
    prompts = []

    # Read each JSON file and extract the 'prompt' attribute from each object in the array
    for file_path in file_paths:
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                data = json.load(f)
                if isinstance(data, list):
                    for item in data:
                        if 'prompt' in item:
                            prompts.append(item['prompt'])
                        else:
                            print(f'Warning: No "prompt" key found in an item in file: {file_path}')
                else:
                    print(f'Warning: File {file_path} does not contain a JSON array')

        except Exception as e:
            print(f'Error reading file {file_path}: {e}')

    # Write all prompts to a JSON file
    combined_json = {
        'prompts': prompts
    }
    with open(output_json_file, 'w', encoding='utf-8') as json_file:
        json.dump(combined_json, json_file, indent=4)

    # Write all prompts to a text file
    with open(output_txt_file, 'w', encoding='utf-8') as txt_file:
        txt_file.write("\n".join(prompts))

                 
file_paths_sound = ['/Users/mimi/Desktop/internship/study/output/2/sound/sound 2.json',
              '/Users/mimi/Desktop/internship/study/output/7/sound/sound 7.json',
              '/Users/mimi/Desktop/internship/study/output/16/sound/sound 16.json',
              '/Users/mimi/Desktop/internship/study/output/23/sound/sound 23.json'] 
output_json_file_sound = '/Users/mimi/Desktop/internship/study/output/combined_prompts_sound.json'
output_txt_file_sound = '/Users/mimi/Desktop/internship/study/output/combined_prompts_sound.txt'
file_paths_story = ['/Users/mimi/Desktop/internship/study/output/2/story/story 2.json',
              '/Users/mimi/Desktop/internship/study/output/7/story/story 7.json',
              '/Users/mimi/Desktop/internship/study/output/16/story/story 16.json',
              '/Users/mimi/Desktop/internship/study/output/23/story/story 23.json'] 
output_json_file_story = '/Users/mimi/Desktop/internship/study/output/combined_prompts_story.json'
output_txt_file_story = '/Users/mimi/Desktop/internship/study/output/combined_prompts_story.txt'
combine_json_files(file_paths_sound, output_json_file_sound, output_txt_file_sound)
combine_json_files(file_paths_story, output_json_file_story, output_txt_file_story)
```
